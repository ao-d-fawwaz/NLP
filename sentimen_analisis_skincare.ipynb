```python
import torch
import seaborn as sns
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud
```

```python
from tqdm.notebook import tqdm
```

```python
from transformers import pipeline
```

```python
pipe=pipeline("sentiment-analysis")
```

```python
pipe("fucking crazy")
```

```python
df=pd.read_csv('/content/Ulta_Skincare_Reviews.csv')
df.head()
```

```python
reviews = df['Review_Text'].fillna('').apply(str).tolist()
```

```python
result_review = {}

# Loop through each review in the list
for i, x in tqdm(enumerate(reviews),total=len(reviews), desc="Processing Reviews"):
    # Apply the sentiment analysis pipeline to the current review
    result = pipe(x)
    # Store the result in the dictionary with the review index as the key
    result_review[i] = result
```

```python
for index, result in tqdm(result_review.items()):
    # Print the index and the result
    print(f"Review {index}: {result}")
```

```python
sentiment_labels = []
sentiment_scores = []

# Iterate over the result_review dictionary to extract labels and scores
for index, result_list in tqdm(result_review.items()):
    # Assuming each result_list contains one result dictionary
    result = result_list[0]  # Extract the first (and only) item in the list
    sentiment_labels.append(result['label'])
    sentiment_scores.append(result['score'])

# Add these lists as new columns to the DataFrame
df['Sentiment_Label'] = sentiment_labels
df['Sentiment_Score'] = sentiment_scores
```
